온도 컬럼(start_temp/start_fellslike/destin_temp/destin_feelslike)는 섭씨온도토 변환을 통해 음수->양수로 전처리


그밖의 제거 대상 전처리는 IQR 사분위수로 통합
->이상치가 극단값이며 각각의 컬럼 특성을 고려할때 제거해도 영향이 없을 것이라 판단

```python
# 이상치 제거 대상
continuous_cols = [
    'travelDurationMinutes',
    'start_windgust',
    'start_windspeed',
    'start_humidity',
    'destin_windgust',
    'destin_windspeed',
    'destin_humidity',
]

# 1. 데이터 불러오기 (샘플 50,000개만)
df = pd.read_csv("최종(전처리 후).csv")

df['start_temp'] = df['start_temp'] * 9/5 + 32
df['start_feelslike'] = df['start_feelslike'] * 9/5 + 32

df['destin_temp'] = df['start_temp'] * 9/5 + 32
df['destin_feelslike'] = df['start_feelslike'] * 9/5 + 32
# 2. IQR 사분위수로 
Q1 = df[continuous_cols].quantile(0.25)
Q3 = df[continuous_cols].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 3. 이상치 제거 마스크 적용
mask = ~((df[continuous_cols] < lower_bound) | (df[continuous_cols] > upper_bound)).any(axis=1)
df_cleaned = df[mask].copy()

# 4. 결과 요약 출력
print(f"제거 전 샘플 수: {len(df)}")
print(f"제거 후 샘플 수: {len(df_cleaned)}")
print(f"제거된 샘플 수: {len(df) - len(df_cleaned)}")

df_cleaned.to_csv("찐최종(이상치 제거).csv", index=False)
```

box cox  변환을 이용하여 baseFare 변환

종속변수에 적용하는 부분이고, 독립변수들은 적용 대상이 아님

```python
# baseFare > 0 필터링
df_boxcox = df_cleaned[df_cleaned['baseFare'] > 0].copy()

# Box-Cox 변환 적용 -> y값은 이제 boxcox_baseFare를 쓸거임.
df_boxcox['boxcox_baseFare'], fitted_lambda = boxcox(df_boxcox['baseFare'])

# 시각화
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_boxcox['boxcox_baseFare'], kde=True, bins=50)
plt.title(f"Box-Cox Transformed baseFare (λ={fitted_lambda:.3f})")

plt.subplot(1, 2, 2)
sns.boxplot(x=df_boxcox['boxcox_baseFare'])
plt.title("Box-Cox Boxplot")

plt.tight_layout()
plt.show()
```

원래 데이터프레임에서 종속변수 y를 예측하기 위하여 사용되는 독립변수들만 뽑기위한 과정

필요없는거 지우고, 필요한 것들만 남기는 과정임

```python
drop_cols = ['baseFare', 'boxcox_baseFare']

# 존재하는 컬럼만 제거
existing = [col for col in drop_cols if col in df_boxcox.columns]
X = df_boxcox.drop(columns=existing)

# 종속변수는 제거 전 데이터에서 꺼내야 함
y = df_boxcox['boxcox_baseFare']

print("제거 전 X shape:", df_boxcox.shape)
print("제거 후 X shape:", X.shape)
print("X.columns에 제거 대상 있는지 확인:", set(drop_cols) & set(X.columns))  # 교집합 없어야 정상
```

```python
# 2. 정규화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. 데이터 분할
X_train, X_valid, y_train, y_valid = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 4. 파라미터 그리드 수동 정의
n_estimators_list = [100, 300, 500]
max_depth_list = [10, 20, 30]
min_samples_split_list = [2, 5, 10]

# 5. 수동 그리드 탐색 + tqdm 진행 표시
results = []
param_combinations = [
    (n, d, s)
    for n in n_estimators_list
    for d in max_depth_list
    for s in min_samples_split_list
]

for n_estimators, max_depth, min_samples_split in tqdm(param_combinations, desc="Grid Search 진행 중"):
    model = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        random_state=42,
        n_jobs=-1
    )
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    mae = mean_absolute_error(y_valid, preds)
    results.append({
        'n_estimators': n_estimators,
        'max_depth': max_depth,
        'min_samples_split': min_samples_split,
        'MAE': mae
    })

# 6. 결과 정리
results_df = pd.DataFrame(results)
best_params = results_df.sort_values(by='MAE').iloc[0]

print("\n 최적 파라미터 조합:")        


   


print(best_params)

```

최적 파라미터를 기준으로 이상치 처리 전 데이터 rf 학습

```python
# 1. 데이터 불러오기
df = pd.read_csv("최종(전처리 후).csv")

# 2. Box-Cox 변환
df = df[df['baseFare'] > 0].copy()
df['boxcox_baseFare'], fitted_lambda = boxcox(df['baseFare'])

# 3. X, y 정의
X = df.drop(columns=['baseFare', 'boxcox_baseFare'])
y = df['boxcox_baseFare']
feature_names = X.columns

# 4. 정규화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. 데이터 분할 (7:1:2 비율)
X_temp, X_test, y_temp, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_temp, y_temp, test_size=0.125, random_state=42)   


# 6. 랜덤 포레스트 모델 정의
rf_model = RandomForestRegressor(
    n_estimators=500,
    max_depth=30,
    oob_score=True,
    min_samples_split=2,
    random_state=42,
    n_jobs=-1
)


# 7. K-Fold 교차검증 (3-Fold)
kf = KFold(n_splits=3, shuffle=True, random_state=42)
r2_scores, mse_scores, mape_scores = [], [], []

for train_idx, val_idx in kf.split(X_train):
    X_tr, X_val = X_train[train_idx], X_train[val_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

    rf_model.fit(X_tr, y_tr)
    y_val_pred = rf_model.predict(X_val)

    r2_scores.append(r2_score(y_val, y_val_pred))
    mse_scores.append(mean_squared_error(y_val, y_val_pred))
    mape_scores.append(mean_absolute_percentage_error(
        inv_boxcox(y_val, fitted_lambda),
        inv_boxcox(y_val_pred, fitted_lambda)
    ))

# 8. 최종 학습
rf_model.fit(X_train, y_train)

# 9. 테스트 예측 및 역변환
y_test_pred = rf_model.predict(X_test)
y_test_original = inv_boxcox(y_test, fitted_lambda)
y_test_pred_original = inv_boxcox(y_test_pred, fitted_lambda)

# 10. 평가 출력
print("\n Random Forest 평가 결과:")
print(f"OOB R²: {rf_model.oob_score_:.4f}")
print(f"Test MSE (Box-Cox): {mean_squared_error(y_test, y_test_pred):.4f}")
print(f"Test R² (Box-Cox): {r2_score(y_test, y_test_pred):.4f}")
print(f"Test MSE (baseFare): {mean_squared_error(y_test_original, y_test_pred_original):.4f}")
print(f"Test R² (baseFare): {r2_score(y_test_original, y_test_pred_original):.4f}")
print(f"Test MAPE (baseFare): {mean_absolute_percentage_error(y_test_original, y_test_pred_original):.4f}")
print(f"CV R² 평균: {np.mean(r2_scores):.4f}")
print(f"CV MSE 평균: {np.mean(mse_scores):.4f}")
print(f"CV MAPE 평균: {np.mean(mape_scores):.4f}")

# 11. 예측 vs 실제 시각화
plt.figure(figsize=(7, 7))
plt.scatter(y_test_original, y_test_pred_original, alpha=0.3, color='dodgerblue', edgecolor='k')
max_val = max(max(y_test_original), max(y_test_pred_original))
plt.plot([0, max_val], [0, max_val], 'r--', label='Perfect Prediction')
plt.xlabel('Actual baseFare')
plt.ylabel('Predicted baseFare')
plt.title('Predicted vs Actual baseFare (Test Set)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

이상치 처리 후 데이터 학습

```python
# 1. 데이터 불러오기
df = pd.read_csv("찐최종(이상치 제거).csv")

# 2. Box-Cox 변환
df = df[df['baseFare'] > 0].copy()
df['boxcox_baseFare'], fitted_lambda = boxcox(df['baseFare'])

# 3. X, y 정의
X = df.drop(columns=['baseFare', 'boxcox_baseFare'])
y = df['boxcox_baseFare']
feature_names = X.columns

# 4. 정규화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. 데이터 분할 (7:1:2 비율)
X_temp, X_test, y_temp, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_temp, y_temp, test_size=0.125, random_state=42)   


# 6. 랜덤 포레스트 모델 정의
rf_model_2 = RandomForestRegressor(
    n_estimators=500,
    max_depth=30,
    oob_score=True,
    min_samples_split=2,
    random_state=42,
    n_jobs=-1
)


# 7. K-Fold 교차검증 (3-Fold)
kf = KFold(n_splits=3, shuffle=True, random_state=42)
r2_scores, mse_scores, mape_scores = [], [], []

print("K-Fold 학습 진행 중...")
for train_idx, val_idx in tqdm(kf.split(X_train), total=kf.get_n_splits(), desc="K-Fold CV"):
    X_tr, X_val = X_train[train_idx], X_train[val_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

    rf_model_2.fit(X_tr, y_tr)
    y_val_pred = rf_model_2.predict(X_val)

    r2_scores.append(r2_score(y_val, y_val_pred))
    mse_scores.append(mean_squared_error(y_val, y_val_pred))
    mape_scores.append(mean_absolute_percentage_error(
        inv_boxcox(y_val, fitted_lambda),
        inv_boxcox(y_val_pred, fitted_lambda)
    ))

# 8. 최종 학습
print("전체 훈련 데이터로 최종 학습 중...")
with tqdm(total=1, desc="Final Training") as pbar:
    rf_model_2.fit(X_train, y_train)
    pbar.update(1)

# 9. 테스트 예측 및 역변환
y_test_pred = rf_model_2.predict(X_test)
y_test_original = inv_boxcox(y_test, fitted_lambda)
y_test_pred_original = inv_boxcox(y_test_pred, fitted_lambda)

# 10. 평가 출력
print("\n Random Forest 평가 결과:")
print(f"OOB R²: {rf_model_2.oob_score_:.4f}")
print(f"Test MSE (Box-Cox): {mean_squared_error(y_test, y_test_pred):.4f}")
print(f"Test R² (Box-Cox): {r2_score(y_test, y_test_pred):.4f}")
print(f"Test MSE (baseFare): {mean_squared_error(y_test_original, y_test_pred_original):.4f}")
print(f"Test R² (baseFare): {r2_score(y_test_original, y_test_pred_original):.4f}")
print(f"Test MAPE (baseFare): {mean_absolute_percentage_error(y_test_original, y_test_pred_original):.4f}")
print(f"CV R² 평균: {np.mean(r2_scores):.4f}")
print(f"CV MSE 평균: {np.mean(mse_scores):.4f}")
print(f"CV MAPE 평균: {np.mean(mape_scores):.4f}")

# 11. 예측 vs 실제 시각화
plt.figure(figsize=(7, 7))
plt.scatter(y_test_original, y_test_pred_original, alpha=0.3, color='dodgerblue', edgecolor='k')
max_val = max(max(y_test_original), max(y_test_pred_original))
plt.plot([0, max_val], [0, max_val], 'r--', label='Perfect Prediction')
plt.xlabel('Actual baseFare')
plt.ylabel('Predicted baseFare')
plt.title('Predicted vs Actual baseFare (Test Set)')
plt.legend()
plt.grid(True)




plt.tight_layout()
plt.show()
```

만든 모델이 너무 무거워 shap가 제대로 작동하지 않는 문제 발생 임의의 파라미터로 shap를 통한 xai 별도 생성

```python
# 1. 데이터 불러오기
df = pd.read_csv("찐최종(이상치 제거).csv")

# 2. Box-Cox 변환
df = df[df['baseFare'] > 0].copy()
df['boxcox_baseFare'], fitted_lambda = boxcox(df['baseFare'])

# 3. X, y 정의
X = df.drop(columns=['baseFare', 'boxcox_baseFare'])
y = df['boxcox_baseFare']
feature_names = X.columns

# 4. 정규화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. 데이터 분할 (7:1:2 비율)
X_temp, X_test, y_temp, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_temp, y_temp, test_size=0.125, random_state=42)

# 6. 랜덤 포레스트 모델 정의
rf_sample = RandomForestRegressor(
    n_estimators=100,
    max_depth=5,
    min_samples_split=5,
    min_samples_leaf=2,
    max_features=0.5,
    oob_score=True,
    random_state=42,
    n_jobs=-1
)

# 7. K-Fold 교차검증 (3-Fold)
kf = KFold(n_splits=3, shuffle=True, random_state=42)
r2_scores, mse_scores, mape_scores = [], [], []

for train_idx, val_idx in kf.split(X_train):
    X_tr, X_val = X_train[train_idx], X_train[val_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

    rf_sample.fit(X_tr, y_tr)
    y_val_pred = rf_sample.predict(X_val)

    r2_scores.append(r2_score(y_val, y_val_pred))
    mse_scores.append(mean_squared_error(y_val, y_val_pred))
    mape_scores.append(mean_absolute_percentage_error(
        inv_boxcox(y_val, fitted_lambda),
        inv_boxcox(y_val_pred, fitted_lambda)
    ))

# 8. 최종 학습
rf_sample.fit(X_train, y_train)

# 9. 테스트 예측 및 역변환
y_test_pred = rf_sample.predict(X_test)
y_test_original = inv_boxcox(y_test, fitted_lambda)
y_test_pred_original = inv_boxcox(y_test_pred, fitted_lambda)

# 1. 샘플 100개 DataFrame으로 변환
X_sample_df = pd.DataFrame(X_train[:100], columns=X.columns)

# 2. SHAP explainer 생성
explainer = shap.TreeExplainer(
    rf_sample,
    feature_perturbation="tree_path_dependent",
    model_output="raw",
    approximate=True
)

# 3. SHAP 값 계산 (진행률 표시)
shap_values_list = []
for i in tqdm(range(len(X_sample_df)), desc="SHAP 진행"):
    shap_val = explainer.shap_values(X_sample_df.iloc[i])
    shap_values_list.append(shap_val)

# 4. numpy 배열로 변환
shap_values = np.array(shap_values_list)

# 5. 전체 피처 시각화
plt.title("SHAP Feature Impact (All Features)")
shap.summary_plot(shap_values, X_sample_df, feature_names=X_sample_df.columns)
```

SHAP 시각화 기반 과적합 위험 혹은 필요없는 변수 3개 제거

1. start_uvindex	SHAP값 매우 낮고 분산 거의 없음 (기여도 ↓)
2. destin_uvindex	상기 변수와 유사, 정보 중복 가능성 큼
3. start_feelslike	start_temp와 중복, SHAP값 낮음

```python
# 1. 데이터 불러오기
df = pd.read_csv("찐최종(이상치 제거).csv")

# 2. Box-Cox 변환
df = df[df['baseFare'] > 0].copy()
df['boxcox_baseFare'], fitted_lambda = boxcox(df['baseFare'])

# 3. X, y 정의
X = df.drop(columns=['baseFare', 'boxcox_baseFare'])

# SHAP 기준으로 제거할 컬럼 정의
drop_vars = [
    "start_uvindex",
    "destin_uvindex",
    "start_feelslike"
]
X = X.drop(columns=drop_vars) 

y = df['boxcox_baseFare']
feature_names = X.columns


# 4. 정규화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. 데이터 분할 (7:1:2 비율)
X_temp, X_test, y_temp, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_temp, y_temp, test_size=0.125, random_state=42)   


# 6. 랜덤 포레스트 모델 정의
rf_final = RandomForestRegressor(
    n_estimators=500,
    max_depth=30,
    oob_score=True,
    min_samples_split=2,
    random_state=42,
    n_jobs=-1
)


# 7. K-Fold 교차검증 (3-Fold)
kf = KFold(n_splits=3, shuffle=True, random_state=42)
r2_scores, mse_scores, mape_scores = [], [], []

print("K-Fold 학습 진행 중...")
for train_idx, val_idx in tqdm(kf.split(X_train), total=kf.get_n_splits(), desc="K-Fold CV"):
    X_tr, X_val = X_train[train_idx], X_train[val_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

    rf_final.fit(X_tr, y_tr)
    y_val_pred = rf_final.predict(X_val)

    r2_scores.append(r2_score(y_val, y_val_pred))
    mse_scores.append(mean_squared_error(y_val, y_val_pred))
    mape_scores.append(mean_absolute_percentage_error(
        inv_boxcox(y_val, fitted_lambda),
        inv_boxcox(y_val_pred, fitted_lambda)
    ))

# 8. 최종 학습
print("전체 훈련 데이터로 최종 학습 중...")
with tqdm(total=1, desc="Final Training") as pbar:
    rf_final.fit(X_train, y_train)
    pbar.update(1)

# 9. 테스트 예측 및 역변환
y_test_pred = rf_final.predict(X_test)
y_test_original = inv_boxcox(y_test, fitted_lambda)
y_test_pred_original = inv_boxcox(y_test_pred, fitted_lambda)

# 10. 평가 출력
print("\n Random Forest 평가 결과:")
print(f"OOB R²: {rf_final.oob_score_:.4f}")
print(f"Test MSE (Box-Cox): {mean_squared_error(y_test, y_test_pred):.4f}")
print(f"Test R² (Box-Cox): {r2_score(y_test, y_test_pred):.4f}")
print(f"Test MSE (baseFare): {mean_squared_error(y_test_original, y_test_pred_original):.4f}")
print(f"Test R² (baseFare): {r2_score(y_test_original, y_test_pred_original):.4f}")
print(f"Test MAPE (baseFare): {mean_absolute_percentage_error(y_test_original, y_test_pred_original):.4f}")
print(f"CV R² 평균: {np.mean(r2_scores):.4f}")
print(f"CV MSE 평균: {np.mean(mse_scores):.4f}")
print(f"CV MAPE 평균: {np.mean(mape_scores):.4f}")

# 11. 예측 vs 실제 시각화
plt.figure(figsize=(7, 7))
plt.scatter(y_test_original, y_test_pred_original, alpha=0.3, color='dodgerblue', edgecolor='k')
max_val = max(max(y_test_original), max(y_test_pred_original))
plt.plot([0, max_val], [0, max_val], 'r--', label='Perfect Prediction')
plt.xlabel('Actual baseFare')
plt.ylabel('Predicted baseFare')
plt.title('Predicted vs Actual baseFare (Test Set)')
plt.legend()
plt.grid(True)




plt.tight_layout()
plt.show()
```
